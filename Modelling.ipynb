### Baseline Model

Using the initial features, create a baseline model to further improve upon.
Use sklearn pipeline in order to maintain consistency in the data processing stages.

**Pipeline Steps**

- Categorical variables
  - Replace missing values with most frequent value
  - One hot encode categorical variables
- Numerical variables
  - Replace missing values with median value
  - scale data using the standard scaler
- Model
  - Logistic regression model with default parameters

```python
cat_pipe = Pipeline([
    ('cat_imp', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(sparse_output=True, categories='auto', handle_unknown='ignore')),
])
ord_pipe = Pipeline([
    ('encoder', OrdinalEncoder()),
])
num_pipe = Pipeline([
    ('num_imp', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler()),
])

base_ct = ColumnTransformer([
    ('cat', cat_pipe, cat_cols),
    ('ord', ord_pipe, ord_cols),
    ('num', num_pipe, num_cols)
])
baseline = Pipeline(steps=[
    ('ct', base_ct),
    ('cls', LogisticRegression())
])
baseline
```

```python
baseline.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report,confusion_matrix
```

```python
cat_lr[target].classes_
```

```python
def evaluate(y_pred, y_test, average='macro'):
    print('Accuracy score: ', accuracy_score(y_pred, y_test))
    print('Precision score: ', precision_score(y_pred, y_test, average=average))
    print('Recall score: ', recall_score(y_pred, y_test, average=average))
    print('F1 score: ', f1_score(y_pred, y_test, average=average))
    print(classification_report(y_test, y_pred))
```

```python
y_pred = baseline.predict(X_test)
evaluate(y_pred, y_test)
```

**Evaluation Metrics**

- Accuracy: proportion of correctly classified instances out of the total instances.
- Precision: Precision indicates the ability of the classifier not to label a negative sample as positive. Important when we want to minimize low revenue.
- Recall: Recall indicates the ability of the classifier to find all the positive samples.  important when we want to minimize missed high revenue movies.
- F1-Score: It provides a balance between precision and recall.
- Classification report: helps us understand the performance of the model across different classes and identify any class-specific issues.

```python
feature_df.date_time = pd.to_datetime(feature_df.date_time)
```

```python
for idx, row in tqdm(feature_df.sort_values(['date_time']).iterrows()):
    temp_df = feature_df[(feature_df.date_time<row['date_time'])&(feature_df.customer_id==row['customer_id'])]
    n_temp = len(temp_df)
    
    feature_df.loc[idx, 'prev_count'] = n_temp
    min_, max_, mean_ = 0, 0, 0
    last_1_usd, last_2_usd = 0, 0
    dif_days = 0
    last_1_pt, last_2_pt = 99, 99
    last_1_ct, last_2_ct = 99, 99
    last_1_st = 99
    
    if n_temp > 0:
        # print(row)
        # print(temp_df)

        min_ = temp_df.total_usd.min()
        max_ = temp_df.total_usd.max() 
        mean_ = temp_df.total_usd.mean()

        last_1_usd = temp_df.total_usd.iloc[0]
        last_2_usd = temp_df.total_usd.iloc[1] if n_temp > 1 else 0

        last_1_pt = temp_df.payment_type.iloc[0]
        last_2_pt = temp_df.payment_type.iloc[1] if n_temp > 1 else 99

        last_1_ct = temp_df.country.iloc[0]
        last_2_ct = temp_df.country.iloc[1] if n_temp > 1 else 99

        last_1_st = temp_df.status.iloc[0]

        dif_days = (row['date_time'] - temp_df.date_time.iloc[0]).days

    feature_df.loc[idx, 'min_spend'] = min_
    feature_df.loc[idx, 'max_spend'] = max_
    feature_df.loc[idx, 'mean_spend'] = mean_
    feature_df.loc[idx, 'last_1_usd'] = last_1_usd
    feature_df.loc[idx, 'last_2_usd'] = last_2_usd
    feature_df.loc[idx, 'last_1_pt'] = last_1_pt
    feature_df.loc[idx, 'last_2_pt'] = last_2_pt
    feature_df.loc[idx, 'last_1_ct'] = last_1_ct
    feature_df.loc[idx, 'last_2_ct'] = last_2_ct
    feature_df.loc[idx, 'last_1_st'] = last_1_st
    feature_df.loc[idx, 'dif_days_customer'] = dif_days
```

#### Stacking ensemble

```python
space={
    # SVC
    'svc_gamma' : hp.choice('svc_gamma', ['auto','scale']),
    'C' : hp.uniform('C',0,10),

    # RF
    'criterion' : hp.choice('criterion', ['entropy','gini']),
    'rf_max_depth' : hp.choice('rf_max_depth', np.arange(10, 1200, dtype=int)),
    'rf_max_features' : hp.choice('rf_max_features',['sqrt','log2',None]),
    'min_samples_leaf' : hp.uniform('min_samples_leaf',0,0.5),
    'min_samples_split' : hp.uniform('min_samples_split',0,1),
    'rf_n_estimators': hp.choice('rf_n_estimators',[10,50,350,550,750,1200,1300,1500]),

    # XGB
    'xgb_max_depth': hp.quniform("xgb_max_depth", 3, 18, 1),
    'xgb_gamma': hp.uniform ('xgb_gamma', 1,9),
    'reg_alpha' : hp.quniform('reg_alpha', 40,180,1),
    'reg_lambda' : hp.uniform('reg_lambda', 0,1),
    'colsample_bytree' : hp.uniform('colsample_bytree', 0.5,1),
    'min_child_weight' : hp.quniform('min_child_weight', 0, 10, 1),
    'xgb_n_estimators': hp.choice('xgb_n_estimators',[10,50,350,550,750,1200,1300,1500]),

    # GB
    'gb_max_features' : hp.choice('gb_max_features',['sqrt','log2',None]),
    'gb_n_estimators': hp.choice('gb_n_estimators',[1, 3, 5, 10,50,350,550,750,1200]),

    # other
    'cat_knn_neighbors': hp.choice('cat_knn_neighbors', [2,5,7,10]),
    'num_knn_neighbors': hp.choice('num_knn_neighbors', [2,5,7,10]),
    'scaler': hp.choice('scaler', [StandardScaler, MinMaxScaler]),
}
```

```python
def objective(space):
    # models
    svc_ = SVC(
        gamma = space['svc_gamma'],
        C = space['C'],
    )

    rf_ = RandomForestClassifier(
        criterion = space['criterion'],
        max_depth = space['rf_max_depth'],
        max_features = space['rf_max_features'],
        min_samples_leaf = space['min_samples_leaf'],
        min_samples_split = space['min_samples_split'],
        n_estimators = space['rf_n_estimators']
    )

    xgb_ = XGBClassifier(
        n_estimators =space['xgb_n_estimators'], max_depth = int(space['xgb_max_depth']), 
        gamma = space['xgb_gamma'], reg_alpha = int(space['reg_alpha']),
        min_child_weight=int(space['min_child_weight']), colsample_bytree=int(space['colsample_bytree'])
    )

    final_estimator = GradientBoostingClassifier(
        n_estimators=space['gb_n_estimators'], max_features=space['gb_max_features'],
    )
    
    cat_pipe = Pipeline([
        ('cat_imp', KNNImputer(n_neighbors=space['cat_knn_neighbors'])),
        ('encoder', OneHotEncoder(sparse_output=False, categories='auto', handle_unknown='ignore')),
    ])
    num_pipe = Pipeline([
        ('num_imp', KNNImputer(n_neighbors=space['num_knn_neighbors'])),
        ('scaler', space['scaler']()),
    ])
    multi_pipe = Pipeline([
        ('multi_encoder', MultiHotEncoder()),
    ])
    title_pipe = Pipeline([
        ('title_encoder', CountVectorizer()),
    ])
    
    base_ct = ColumnTransformer([
        ('cat', cat_pipe, cat_cols),
        ('multi', multi_pipe, multi_cols),
        ('num', num_pipe, num_cols),
        # ('title', title_pipe, 'title')
    ])
    
    baseline_stack = Pipeline(steps=[
        ('ct', base_ct),
        ('cls', StackingClassifier(
            estimators=[
                ('rf', rf_), 
                ('xgb', xgb_), 
                ('svc', svc_)
            ],
            final_estimator=final_estimator
        ))
    ])
    baseline_stack
    
    accuracy = cross_val_score(baseline_stack, X_train,y_train,cv=10, scoring='accuracy').mean()
    print("Mean accuracy: ", accuracy)
    return {'loss' : -accuracy , 'status' : STATUS_OK}
```

```python
trials = Trials()
best = fmin(
    fn = objective,
    space=space,
    algo= tpe.suggest,
    max_evals=100,
    trials=trials
)
```

### Ensemble

Combination of different models to produce a final prediction

- Voting Classifier
- Stacking Classifier

**Voting Classifier**

Using a majority vote on the outputs of other models

```python
from sklearn.ensemble import VotingClassifier, StackingClassifier, AdaBoostClassifier
```

```python
cat_pipe = Pipeline([
    ('cat_imp', KNNImputer(n_neighbors=5)),
    ('encoder', OneHotEncoder(sparse_output=False, categories='auto', handle_unknown='ignore')),
])
num_pipe = Pipeline([
    ('num_imp', KNNImputer(n_neighbors=5)),
    ('scaler', StandardScaler()),
])
ord_pipe = Pipeline([
    ('encoder', OrdinalEncoder()),
])

base_ct = ColumnTransformer([
    ('cat', cat_pipe, cat_cols),
    ('ord', ord_pipe, ord_cols),
    ('num', num_pipe, num_cols),
])
baseline_vote = Pipeline(steps=[
    ('ct', base_ct),
    ('vote', VotingClassifier(        # Voting classifier
        estimators=[
            ('rf', RandomForestClassifier(n_estimators=1000, criterion='entropy')), 
            ('xgb', XGBClassifier(n_estimators=1000, max_depth = 1000)), 
            ('dt', DecisionTreeClassifier(max_depth=1000)),
            ('lr', LogisticRegression(max_iter=1000)),
            ('svc', SVC(probability=True)),
        ],
        voting='soft'
    ))
])
baseline_vote
```
